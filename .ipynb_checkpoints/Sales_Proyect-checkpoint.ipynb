## ***************************************************************************************************
## Lets import  pandas, this library will help to convert our csv documents to DataFrames for analysis
## For visualizations, I added the Matplotlib & Seaborns libraries
## ***************************************************************************************************


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import numpy as np
```

```python
# Read the first csv, convert it to DataFrame and show only the first 5 rows
df1 = pd.read_csv(r"data/Datos_Ventas_Tienda.csv")
df1.head(5)
```

```python
# Reading the 2nd csv and show the first 5 rows 
df2 = pd.read_csv(r"data/Datos_Ventas_Tienda2.csv")
df2.head()
```

```python
# Concat both csv documents and change column names to English

df_combined = pd.concat([df1, df2], ignore_index=True)
names2 = ["Date", "Product","Quantity", "Price Per Unit", "Total"]
df_combined.columns = names2
df_combined.head()
```

# Before working on the DataFrame, lets see the right Dtype in all columns

```python
#Just to verify that all the columns have the proper Dtype, let's gather the information with .info()

df_combined.info()
```

```python
## Column "Date" shows a DType "object" lets Change it to DateTime

df_combined["Date"] = pd.to_datetime(df_combined["Date"])
df_combined.info()
```

## Once our data is clean and clear, lets answer our first bussines question

## 1. Which product generated the highest total sales, and what was its specific monetary contribution to the company?


```python
#Adding a new column to our DataFrame called "Month" when I will extract only the month
df_combined ["Month"] = df_combined["Date"].dt.month_name()

#Grouping by product and summing the Quantity and Total sales, then sort values by the sum of Total 
df_product_sales = df_combined.groupby("Product")[["Quantity" , "Total"]].sum().sort_values(by = "Total"  , ascending = False)
df_product_sales = df_product_sales.reset_index()
df_product_sales



```

## 2. How does the sales volume of each product in December compare to our peak-performing month, and what is the percentage variance between these two periods?"

```python
#Let's group by "Month" and "Product" but only when "Month== "December" only to see the Total sales for each product on this specific time

total_december = df_combined[df_combined["Month"] == "December"].groupby(["Month" , "Product"])["Total"].sum().sort_values( ascending =False)
total_december


```

```python
#Now it's important to check wich is the month with the highest sales 
highest_month = df_combined.groupby("Month")["Total"].sum().idxmax()
print(f"The month with highest sales is {highest_month}")

```

```python
# Since we know that the highest month is "August", lets see the "Products" and "Total" sales on this month (as I did before with "December")
 
total_august = df_combined[df_combined["Month"] == "August"].groupby(["Month" , "Product"])["Total"].sum().sort_values( ascending =False)
print(f"{total_august}\n")
print(total_december)
```

```python
print(f"The sales in December are lower than August, none of the products reach the sales in August is recommended use more marketing to increase the sales on this special month")
```

## 3. Which product exhibited the lowest sales volume, and during which month did this occur?


```python
#To see the lowest prodcuts sales, lets grouo by "Product" or "Month" and summing the "Total" sales
df_lowest = df_combined.groupby(["Product","Month"])["Total"].sum().sort_values(ascending =True)
#type(df_lowest)
print(f"The lowest product is {df_lowest.index[1][0]} and had a total sale : {df_lowest.iloc[0]}")
```

## 4. Is there a statistically significant correlation between bi-weekly pay cycles and sales volume fluctuations, or does the impact remain negligible?

```python
#To verify this data, let's create a new "Column" called "Payday" to our combined DataFrame and select only the days that are
# included on the days 14,15,16 of each month

df_combined["Payday"] = df_combined["Date"].dt.day.isin([14,15,16]) #Will provide a boolean 
comparative = df_combined.groupby("Payday")["Total"].mean()
print(f"Normal days ${comparative[False]:.2f}\n")
print(f"Paydays ${comparative[True]:.2f}\n")

impact_in_percent = ((comparative[True] - comparative[False]) / comparative[False]) * 100

print(f"The reflected sales on Paydays have an impact: {impact_in_percent:.2f}%")
```

## The code below shows a bar chart  that indicates  Total sales per each Product


```python


fig, ax = plt.subplots(figsize = (18,12))
fig.set_facecolor("#1e1e1e")
ax.set_facecolor("#2d2d2d")


plt.bar(df_product_sales['Product'], df_product_sales['Total'], label = "Sales", color = "g")


ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))


for i, value in enumerate (df_product_sales['Total']):
    ax.text(i, value + 1, f"${value:,.0f}",
            ha = "center", va = "bottom", fontsize = 30, weight = "bold", color ="white",  #Possition and color
            bbox = dict(facecolor='#27ae60', alpha = 0.5 ,edgecolor = "white", boxstyle='round,pad=0.2'))


plt.title("TOTAL SALES PER YEAR", color = "White", fontsize = 35)
plt.xlabel("\nPRODUCT", c ="#FF3A3A", weight = "bold",  fontsize = 29)
plt.xticks(c = "white", fontsize = 25)
plt.ylabel("SALES (MXN)\n", c = "#1E781E", weight ="bold", fontsize = 29)
plt.yticks(c = "white", fontsize = 25)
plt.grid(True, linestyle= ':', alpha = 0.6)
plt.legend(fontsize = 25)
plt.tight_layout()
```

## Our data identifies "Electronic" as the lead category, consistently maintaining the highest sales volume across almost every month.

## Now, lets see the Electronic Product behavior on each month

```python

#Sorting data by the Product "Electrónic" which is the most sale product, then sort the months from January to December 

df_electronic = df_combined[df_combined["Product"] == "Electrónic"].copy()
months_sorted = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
df_electronic["Month"] = pd.Categorical(df_electronic['Month'], categories = months_sorted, ordered = True)


df_plot = df_electronic.groupby('Month', observed = False)['Total'].sum().reset_index()
```

```python
#Showing the data only for "Electrónic product" during the months before sorted

fig, ax = plt.subplots(figsize=(15, 8))
fig.set_facecolor("#1e1e1e")
ax.set_facecolor("#2d2d2d")

ax.plot(df_plot['Month'], df_plot['Total'], color="g", marker="o", linewidth=3, markersize=10, label="Electronic Sales")

# $ Format
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

for i, value in enumerate(df_plot['Total']):
    if value > 0:
        ax.text(i, value + (value * 0.02), f"${value:,.0f}", 
                ha="center", va="bottom", fontsize=15, weight="bold", color="white",
                bbox=dict(facecolor='#27ae60', alpha=0.7, edgecolor="white", boxstyle='round,pad=0.2'))

# Estética final
plt.title("ELECTRONIC SALES TREND", color="white", weight= "bold", fontsize=25, pad=20)
plt.xticks(color="white", fontsize=12)
plt.xlabel("\nElectronic product per month", c = "#74CCFC",weight ="bold", fontsize = 29)
plt.yticks(color="white", fontsize=12)
plt.ylabel("Sale values (MXN)", c = "#FF9908", fontsize = 29)
ax.grid(True, linestyle=':', alpha=0.3)

plt.tight_layout()
```

## It is critical to ensure optimal inventory levels for this item on August Month, introduce new strategies for the lowest months "March , April and November" and explore horizontal expansion by introducing related accessories or complementary products within this category.

## ***************************************************************************************************************************************************
## 2. How does the sales volume of each product in December compare to our peak-performing month, and what is the percentage variance between these two periods?"

## Although December shows high sales activity, total sales still falls short of our annual peak-performance month. To bridge this gap, a strategic investment in a specialized marketing team is recommended. This team should focus on customer acquisition strategies and seasonal promotions to maximize sales during the holiday period.

## ***************************************************************************************************************************************************


```python
data_aug = total_august.reset_index()
data_aug

data_dec = total_december.reset_index()
data_dec

width = 0.35 

x = np.arange(len(data_aug['Product'])) 

fig, ax = plt.subplots(figsize=(18, 10))
fig.set_facecolor('#1e1e1e')
ax.set_facecolor('#2d2d2d')


ax.tick_params(axis='y', colors='white', labelsize=15)

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

rects1 = ax.bar(x - width/2, data_aug['Total'], width, label='August', color='#3498db')
rects2 = ax.bar(x + width/2, data_dec['Total'], width, label='December', color='#27ae60')


ax.set_xticks(x)
ax.set_xticklabels(data_aug['Product'], color="white", fontsize=15)


def add_labels(rects, bubble_color):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height + (height * 0.02),
                f'${height:,.0f}',
                ha='center', va='bottom', color='white', 
                fontweight='bold', fontsize=12,
                bbox=dict(facecolor=bubble_color, alpha=0.6, edgecolor='none', boxstyle='round,pad=0.2'))

add_labels(rects1, '#3498db')
add_labels(rects2, '#27ae60')

ax.set_ylabel("Total sales (MXN)", color="white",fontsize=20, fontweight="bold", labelpad=20)

plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
    
```


## 3. Which product exhibited the lowest sales volume, and during which month did this occur?

## The "Alimentos" category consistently reports the lowest sales. Given its low market demand and the inherent risk of inventory spoilage (due to perishability), we should evaluate a phased exit strategy or a significant reduction in stock levels to optimize capital allocation.

```python
#Sort data for the highest and the lowest products
data1 = ["Electrónic", "Alimentos"]
months  = df_combined[df_combined["Product"].isin(data1)].groupby([df_combined["Date"].dt.month, "Product"])["Total"].sum().unstack()


#Once this data has been prepared, let's ilustrate the behavior on the next Visualization.

fig, axe = plt.subplots(figsize = (12,7))

fig.set_facecolor('#1e1e1e')
axe.set_facecolor('#2d2d2d')


months.plot(kind='line', ax=axe, marker='o', linewidth=3)
axe.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))


for products in months.columns:
    for x, y in zip(months.index, months[products]):
        if pd.notnull(y):
            axe.annotate(f'${y:,.0f}', (x, y), textcoords="offset points", xytext=(0, 12), 
                         ha='center', fontsize=9, color='white', fontweight='bold',
                         bbox=dict( boxstyle='round,pad=0.3', fc='#3d3d3d', ec='white', alpha=0.8))



plt.title("LOWEST SALES THIS YEAR", c= "white", fontsize = 25)
plt.xlabel("Month Number", color='white', fontsize=12)
plt.ylabel("Total Sales", color='white', fontsize=12)


axe.tick_params(colors = 'white')
for spine in axe.spines.values():
    spine.set_color('white')

#Configure legends
legend = axe.legend(facecolor='#1e1e1e', edgecolor='white', labelcolor='white')
axe.grid(color='gray', linestyle='--', alpha=0.3)
```

## 4 Is there a statistically significant correlation between bi-weekly pay cycles and sales volume fluctuations, or does the impact remain negligible?

## Analysis indicates that customer liquidity is a primary driver of sales. To capitalize on this, we must ensure peak inventory availability 48 hours prior to paydays and schedule full staff coverage to handle the increased transaction volume.

```python
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(8, 6))

fig.set_facecolor('#1e1e1e')
ax.set_facecolor("#2d2d2d")

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))


ax = sns.barplot(x= comparative.index, y=comparative.values, hue=comparative.index, palette="viridis", legend= False)

plt.xticks(ticks=[0, 1], labels=['Normal Days', 'Paydays'], color ="white")
plt.yticks(color ="white")
plt.title('Total Average Sales: Normal days vs Paydays \n\n', fontsize=14, color = "white")
plt.ylabel('Total Average Sales (MXN)', color = "white")
plt.xlabel('Days', color = "white")

for p in ax.patches:
    ax.annotate(f'${p.get_height():.2f}',  (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha = 'center', va = 'center', xytext = (0, 9), textcoords = 'offset points', fontsize=12, fontweight='bold', color = "white")
plt.tight_layout()
plt.show()
```
